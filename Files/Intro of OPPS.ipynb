```python
## Abstraction

class Car:
    def __init__(self):
        self.acc = False
        self.brake = False
        self.clutch = False
        self.__model  = 2000

    def start(self):
        self.acc = True
        self.clutch = True
        print("Car is Started...")

    def stop(self):
        self.brake = True
        self.clutch = True
        print("Car is Stop...")

    def get_model(self):
        return self.__model
    def set_model(self, new_model):
        if new_model>=2025:
            return "invalid model"
        else:
            self.__model = new_model
            return "model updated"
```

```python
startCar = Car()
```

```python
startCar.stop()
```

```python
startCar.start()
```
