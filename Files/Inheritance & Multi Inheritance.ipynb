```python
class Animal:
    def speak(self):
        return f"Animal Voices"

class Cat(Animal):
    def meow(self):
        return f"Cat Meow"

class Dog(Animal):
    def bark(self):
        return f"Dog Bark"
```

```python
class Animal:
    def speak(self):
        print(f"Animal Voices")

class Cat(Animal):
    def meow(self):
        return f"Cat Meow"

class Dog(Animal):
    def bark(self):
        return f"Dog Bark"

class Snake(Animal):
    def speak(self):
        super().speak()
        print(f"Snake Voice")
```

```python
animal3 = Snake()
```

```python
animal3.speak()
```

# When a method of Parent class is redefined in child class method overridding : Polymorphsim

```python
class Car:
    wheel =4 
    def __init__(self, make, model, year, color):
        self.make = make
        self.model = model 
        self.year = year
        self.color = color
    def start(self):
        return f"{self.make} {self.model} is started"
    def fill_gas(self):
        return f"{self.make} {self.model} is filling gasoline"
    def stop(self):
        return f"{self.make} {self.model} is stopped"
```

```python
car1 = Car('Honda','Civic',2020, 'Black')
```

```python
Car.wheel
```

```python
car1.wheel
```

```python
Car.wheel = 6
```

```python
car1.wheel
```

```python
car1.price = 10000000
```

```python
class ElectricCar(Car):
    def fill_gas(self):
        return "electic car has no fuel tank"
```

```python
ecar1 = ElectricCar('Honda','Civic',2020, 'Black')
```

```python
ecar1.fill_gas()
```

# Multilevel Inheritance & Multiple Inheritance

```python
class A():
    def __init__(self):
        print('I am class A constructor')
```

```python
class B():
    def __init__(self):
        print('I am class B constructor')
        
```

```python
class B(A):
    def __init__(self):
        print('I am class B constructor')
        super().__init__()
```

```python
class C(B):
    def __init__(self):
        super().__init__()
        print('I am class C constructor')
```

```python
a = A()
```

```python
b = B()
```

```python
c = C()
```

```python
class C(A,B):
    def __init__(self):
       
        print('I am class C constructor')
        super().__init__()
```

```python
c = C()
```
